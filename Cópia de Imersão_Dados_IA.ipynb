## Representação Gráfica do percentual sobre o preço atualizado para real

```python
plt.figure(figsize=(10, 6))
plt.scatter(dados['Percent'], dados['actual price $'], alpha=0.5)
plt.xlabel('Percent')
plt.ylabel('actual price $')
plt.title('Relação entre Percent e actual price $')
plt.grid(True)
plt.show()
```

###Box Plot (Diagrama de Caixa): O box plot é uma representação gráfica que mostra a distribuição de dados, incluindo os quartis e os valores atípicos. Os valores que estão significativamente fora das "caixas" (interquartil) são geralmente considerados outliers.

```python
sns.boxplot(x= dados['Percent'])
plt.title('Box Plot para a coluna Percent')
plt.show()
```

### O Z-Score mede o quão longe um ponto de dados está da média em termos de desvios padrão. Valores de Z-Score muito altos ou baixos podem indicar a presença de outliers. Você pode usar a função zscore do módulo scipy.stats para calcular o Z-Score das colunas relevantes.

```python
from scipy import stats

z_scores = stats.zscore(dados['Percent'])
outliers = (z_scores > 3) | (z_scores < -3)
```

###QR (Intervalo Interquartil): O IQR é a diferença entre o primeiro quartil (25º percentil) e o terceiro quartil (75º percentil) dos dados. Valores fora do intervalo definido por Q1 - 1.5 * IQR e Q3 + 1.5 * IQR são considerados outliers. Você pode calcular o IQR e identificar outliers manualmente.

```python
Q1 = dados['Percent'].quantile(0.25)
Q3 = dados['Percent'].quantile(0.75)
IQR = Q3 - Q1
outliers = (dados['Percent'] < Q1 - 1.5 * IQR) | (dados['Percent'] > Q3 + 1.5 * IQR)
```

###Visualização de Scatter Plot: Plotar um gráfico de dispersão (scatter plot) das colunas em relação uma à outra pode ajudar a identificar visualmente pontos que estão longe da tendência geral dos dados. Pontos isolados podem ser outliers.

```python
plt.scatter(dados['Percent'], dados['actual price $'])
plt.xlabel('Percent')
plt.ylabel('actual price $')
plt.title('Scatter Plot')
plt.show()
```

###Depois de identificar os outliers, você pode optar por tratá-los (removendo, transformando ou substituindo, dependendo do contexto) ou usá-los como informações importantes, dependendo dos objetivos da sua análise. Certifique-se de entender o contexto dos dados e considerar a natureza dos outliers antes de tomar qualquer ação.
